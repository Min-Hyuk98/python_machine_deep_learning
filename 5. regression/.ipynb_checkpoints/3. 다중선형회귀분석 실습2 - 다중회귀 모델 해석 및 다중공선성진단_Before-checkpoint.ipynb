```python
import os
import pandas as pd 
import numpy as np
import statsmodels.api as sm
```

```python
# 현재경로 확인
os.getcwd()
```

```python
# 데이터 불러오기
boston = pd.read_csv("./data/Boston_house.csv")
boston.head()
```

```python
boston_data = boston.drop(['Target'],axis=1)
# boston_data
```

# crim, rm, lstat을 통한 다중 선형 회귀분석

```python
x_data=boston[['CRIM','RM','LSTAT']] ##변수 여러개
target = boston[['Target']]
x_data.head()
```

```python
x_data1 = sm.add_constant(x_data, has_constant='add')
```

```python
multi_model = sm.OLS(target,x_data1)
fitted_multi_model=multi_model.fit()
```

```python
fitted_multi_model.summary()
```

# crim, rm, lstat, b, tax, age, zn, nox, indus 변수를 통한 다중선형회귀분석

```python
## bostan data에서 원하는 변수만 뽑아오기 
x_data2 = boston[['CRIM','RM','LSTAT','B','TAX','AGE','ZN','NOX','INDUS']]
x_data2.head()
```

```python
# 상수항추가
x_data2_ = sm.add_constant(x_data2,has_constant="add")
```

```python
# 회귀모델 적합
multi_model2 = sm.OLS(target, x_data2_)
fitted_multi_model2 = multi_model2.fit()
```

```python
# 결과 출력
fitted_multi_model2.summary()

##위에서 변수 3개만 넣은것과 R-square값이 큰 차이 없다 -> 성능차이가 크지 않다 -> 변수 개수 적은 것이 훨씬 좋다
## [INDUS]의 p-value가 매우 높으므로 이 변수를 제거하고 다시 모델을 돌려볼 필요가 있다
```

```python
# 세변수만 추가한 모델의 회귀 계수 
fitted_multi_model.params
```

```python
# full모델의 회귀계수
fitted_multi_model2.params
```

```python
# base모델과 full모델의 잔차비교 
import matplotlib.pyplot as plt
fitted_multi_model.resid.plot(label="full")
fitted_multi_model2.resid.plot(label="full_add")
plt.legend()

## 큰 차이가 없다
```

# 상관계수/산점도를 통해 다중공선성 확인

```python
# 상관행렬  correlation
x_data2.corr()  # 대칭행렬 형태
```

```python
## 상관행렬 시각화 해서 보기 
import seaborn as sns;
cmap = sns.light_palette("darkgray", as_cmap=True)
sns.heatmap(x_data2.corr(), annot=True, cmap=cmap)
plt.show()
## NOX와 INDUS사이의 correlation이 특히 높다
```

```python
## 변수별 산점도 시각화
sns.pairplot(x_data2)
plt.show()
```

# VIF를 통한 다중공선성 확인 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(x_data2.values, i) for i in range(x_data2.shape[1])]
vif["features"] = x_data2.columns
vif

## VIF가 10 이상이면 다중공선성이라고 진단할 수 있다
## vif가 가장 큰 NOX를 제거해야한다
```

```python
## nox 변수 제거후(X_data3) VIF 확인 

vif = pd.DataFrame()
x_data3= x_data2.drop('NOX',axis=1)
vif["VIF Factor"] = [variance_inflation_factor(
    x_data3.values, i) for i in range(x_data3.shape[1])]
vif["features"] = x_data3.columns
vif

## 아직 RM의 vif가 매우 크다. -> RM 제거
## 사실은, correlation이나 산점도를 볼 때 [RM]은 제거하지 않는게 옳다
```

```python
## RM 변수 제거후(x_data4) VIF 확인 

vif = pd.DataFrame()
x_data4= x_data3.drop('RM',axis=1)
vif["VIF Factor"] = [variance_inflation_factor(
    x_data4.values, i) for i in range(x_data4.shape[1])]
vif["features"] = x_data4.columns
vif

## 여기서 변수 제거 멈추고 모델 적합시켜보자!!
```

```python
# nox 변수 제거한 데이터(x_data3) 상수항 추가 후 회귀 모델 적합
x_data3_ = sm.add_constant(x_data3,has_constant="add")
multi_model3 = sm.OLS(target, x_data3_)
fitted_multi_model3 = multi_model3.fit()


# nox, rm 변수 제거한 데이터(x_data4) 상수항 추가 후 회귀 모델 적합
x_data4_ = sm.add_constant(x_data4,has_constant="add")
multi_model4 = sm.OLS(target, x_data4_)
fitted_multi_model4 = multi_model4.fit()
```

```python
## 회귀모델 결과 비교 
fitted_multi_model3.summary()

## nox제외 전과 R-square 거의 동일 -> nox가 차지하는 변동성이 매우 적었다
## [CRIM] 과 [INDUS]의 p-value가 줄긴 했지만 여전히 높다
```

```python
## 회귀모델 결과 비교 
fitted_multi_model4.summary()

## [RM]변수를 제거했더니 R-square가 많이 낮아졌다 -> [RM]은 차지하는 변동성이 높다 -> 중요한 변수로, 지우면 안됨
```

# 학습 / 검증데이터 분할

```python
from sklearn.model_selection import train_test_split
## 변수 9개 모두 넣음
X = x_data2_
y = target
train_x, test_x, train_y, test_y = train_test_split(X, y, train_size=0.7, test_size=0.3,random_state = 1) 
## random_state=1로 고정시킴 -> 몇번을 실행해도 같은 데이터끼리 분할됨
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape)
train_x.head()
```

```python
# train_x 회귀모델 적합
fit_1 = sm.OLS(train_y, train_x)
fit_1 = fit_1.fit()
```

```python
## 검증데이터에 대한 예측값과 true값 비교 
plt.plot(np.array(fit_1.predict(test_x)),label="pred")
plt.plot(np.array(test_y),label="true")
plt.legend()
plt.show()

## 튀는곳이 조금 있지만 어느정도 맞음
```

```python
## x_data3 학습 검증데이터 분할

X = x_data3_
y = target
train_x3, test_x3, train_y3, test_y3 = train_test_split(X, y, train_size=0.7, test_size=0.3,random_state = 1) 
```

```python
# x_data3 회귀 모델 적합(fit3)

fit_3 = sm.OLS(train_y3, train_x3)
fit_3 = fit_3.fit()
```

```python
## x_data4 학습 검증데이터 분할
X = x_data4_
y = target
train_x4, test_x4, train_y4, test_y4 = train_test_split(X, y, train_size=0.7, test_size=0.3,random_state = 1) 
```

```python
# x_data4 회귀 모델 적합(fit4)

fit_4 = sm.OLS(train_y4, train_x4)
fit_4 = fit_4.fit()
```

```python
## true값과 예측값 비교 
plt.plot(np.array(fit_3.predict(test_x3)),label="pred3") # [NOX] 제거
plt.plot(np.array(fit_4.predict(test_x4)),label="pred4") # [NOX], [RM] 제거
plt.plot(np.array(test_y3),label="true") # random_state고정 -> test_y == test_y3 == test_y4
plt.legend()
plt.show()

##결과적으로 [NOX]만 제거한것이 더 좋음
```

```python
## full모델 추가해서 비교 
plt.plot(np.array(fit_1.predict(test_x)),label="pred")
plt.plot(np.array(fit_3.predict(test_x3)),label="pred_vif3")
plt.plot(np.array(fit_4.predict(test_x4)),label="pred_vif4")
plt.plot(np.array(test_y),label="true")
plt.legend()
plt.show()
```

```python
## 각각의 잔차 그래프
plt.plot(np.array(test_y['Target']-fit_1.predict(test_x)),label="pred_full")
plt.plot(np.array(test_y['Target']-fit_3.predict(test_x3)),label="pred_vif3")
plt.plot(np.array(test_y['Target']-fit_4.predict(test_x4)),label="pred_vif4") 
plt.legend()
plt.show()
 # [NOX], [RM] 제거한것이 잔차가 가장 크다 
```

# MSE를 통한 검증데이터에 대한 성능비교 

```python
from sklearn.metrics import mean_squared_error
```

```python
# crim, rm, lstat, b, tax, age, zn, nox, indus 변수 사용
mean_squared_error(y_true=test_y['Target'], y_pred=fit_1.predict(test_x))
```

```python
# crim, rm, lstat, b, tax, age, zn, indus 변수 사용... [NOX] 제거
mean_squared_error(y_true=test_y['Target'], y_pred=fit_3.predict(test_x3))
```

```python
# crim, lstat, b, tax, age, zn, indus 변수 사용... [NOX], [RM] 제거
mean_squared_error(y_true=test_y['Target'], y_pred=fit_4.predict(test_x4))
```

- [NOX], [RM] 제거는 mse 가장 높으므로 가장 안좋은 모델
- [NOX]만 제거한 모델과 9개 변수 사용한 모델은 mse가 비슷한데,
    - 이때는 변수가 적을수록 좋음 --> [NOX]만 제거한 모델이 셋중에서 가장 좋음
