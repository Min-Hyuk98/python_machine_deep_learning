## 학습목표
 1. matplotlib 모듈 이용하여 그래프 표현하기

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
## 그래프가 jupyter 내부에서 그려지도록 설정
```

### 그래프 데이터 생성

```python
x = np.linspace(0, 10, 11)
y = x**2 + x + 2 + np.random.randn(11)  ## x원소가 11개니까 random.randn(11)
print(x)
print(y)
```

### 그래프 출력하기
 - plot함수 (선 그래프), scatter(점 그래프), hist(히스토그램) 등 사용
   - 함수의 parameter 혹은 plt의 다른 함수로 그래프 형태 및 설정을 변경 가능
   - 기본적으로, x, y에 해당하는 값이 필요

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

### 그래프에 주석 추가 

* **x, y 축 및 타이틀**

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('x-y graph')

plt.plot(x,y)
```

* **grid 추가**

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('x-y graph')

plt.grid(True)

plt.plot(x,y)
```

* **x, y축 범위 지정**

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('x-y graph')
plt.grid(True)

plt.xlim(0,20)
plt.ylim(0,100)

plt.plot(x,y)
```

### plot 함수 parameters
 - 그래프의 형태에 대한 제어 가능
 - [plot함수 도큐먼트](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)

* 그래프의 색상 변경

```python
plt.plot(x,y, 'r')

# r -> red
# b -> blue
# y -> yellow
# k -> black
```

```python
plt.plot(x,y, '#00ffff')
```

* 그래프 선스타일 변경

```python
plt.plot(x,y, '-.')
```

```python
plt.plot(x,y, 'g^') ## green ^모양
```

```python
plt.plot(x,y, 'y:')  # yellow 점선
```

* 그래프 두께 변경
 - linewidth 명시

```python
plt.plot(x,y, 'k:', linewidth=10)
```

* keyword parameter 이용하여 모든 속성 설정
 - https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
 - color
 - linestyle
 - marker
 - markerfacecolor
 - markersize 등등

```python
plt.plot(x,y, color='black', linestyle='-', marker='*', markerfacecolor='red', markersize=15)
```

### subplot으로 여러 그래프 출력하기
 - subplot함수로 구획을 구별하여 각각의 subplot에 그래프 출력

```python
plt.subplot(2,2,1)  ## 2 x 2 구역중 첫번째 (왼쪽 위)
plt.plot(x,y,'r')

plt.subplot(2,2,2)  ## 2 x 2 구역중 두번째 (오른쪽 위)
plt.plot(x,x**2,'g')

plt.subplot(2,2,3)  ## 2 x 2 구역중 세번째 (왼쪽 아래)
plt.plot(x,np.sin(x),'y')

plt.subplot(2,2,4)  ## 2 x 2 구역중 네번째 (오른쪽 아래)
plt.plot(x,np.exp(x),'k')
```

### hist함수 사용
 - histogram 생성
 - bins로 historgram bar 개수 설정

```python
data = np.random.randint(1,100, size=200)
```

```python
plt.hist(data, bins=15, alpha=0.5)  ## bin은 막대 갯수, alpha는 막대 투명도
plt.xlabel('value')
plt.ylabel('number')
plt.grid(True)
```
