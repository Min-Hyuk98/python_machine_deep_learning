## 학습목표
 1. dataframe row 선택하기

```python
import pandas as pd
import numpy as np
```

```python
# data 출처: https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
train_data = pd.read_csv('./train.csv')
train_data.head()
```

#### dataframe slicing
  - dataframe의 경우 기본적으로 [] 연산자가 **column 선택**에 사용
  - 하지만, **slicing은 row 레벨**로 지원

```python
## 3행 ~ 8행을 출력
train_data[3:9]
```

#### row 선택하기
  - Seires의 경우 []로 row 선택이 가능하나, **DataFrame의 경우는 기본적으로 column을 선택하도록 설계**
  - **.loc, .iloc함수**로 row 선택 가능
    - loc - 인덱스 자체를 사용
    - iloc - 0 based index로 사용
    - 이 두 함수는 ,를 사용하여 column 선택도 가능

```python
train_data.info() 

# 총 행이 891개인 것을 알 수 있다
```

```python
# loc와 iloc의 차이를 구분하기 위해 임의로 인덱스를 1000 ~ 1890으로 지정해줌
train_data.index = np.arange(1000,1891)
train_data.head()
```

### loc : 해당 인덱스의 행을 가져옴

```python
train_data.loc[1003]
```

```python
## 여러개 행을 가져올 때는 idx 리스트를 안에 넣음
train_data.loc[ [1003, 1005, 1321] ]
```

### iloc : 지정된 인덱스 관계 없이, 0번째 행부터 시작해서 n번째 행을 가져옴

```python
## iloc을 사용하면, 지정된 인덱스 관계 없이, 0번째 행부터 시작해서 n번째 행을 가져옴
train_data.iloc[ [3,5,321] ]

# train_data.loc[ [1003, 1005, 1321] ] 와 동일한 것을 볼 수 있다
```

## row, coloumn 동시에 사용하기

- loc, iloc 속성을 이용할 때 콤마를 이용해서 둘 다 명시 가능

```python
# df.loc[ [행 인덱스들], [열 인덱스들] ]

train_data.loc[ [1003, 1005, 1321], ['Survived', 'Name', 'Age'] ]
```

```python
# [3,5,321]은 행의 0부터 시작하는 인덱스
# [1,3,5]는 열의 0부터 시작하는 인덱스

train_data.iloc[ [3,5,321], [1,3,5] ]
```
