 1. NaN 처리 방법 이해하기


```python
import pandas as pd
```

```python
# data 출처: https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
train_data = pd.read_csv('./train.csv')
train_data.head()
```

#### NaN 값 확인
 - info함수를 통하여 개수 확인
 - isna함수를 통해 boolean 타입으로 확인

```python
train_data.info()

# 총 891개 데이터중 NaN데이터가 있다
```

```python
train_data.isna().head()

# true인 것이 NaN
```

```python
train_data['Age'].isna().head()
```

#### NaN 처리 방법
 - 데이터에서 삭제
   - dropna 함수 
 - 다른 값으로 치환
   - fillna 함수

* NaN 데이터 삭제하기

```python
train_data.dropna().head()

# NaN 있는 행 전체를 삭제
```

```python
train_data.dropna(subset=['Age']).head()

# 'AGE'열에 NaN있는 행 모두 삭제.. 
# 다른 열에 있는 NaN은 신경 안씀
```

```python
train_data.dropna(axis=1).head()

## axis=1 -> 열을 기준으로 돌아감
## NaN이 하나라도 있는 열을 모두 삭제한다
```

* NaN 값 대체하기
 - 평균으로 대체하기
 - 생존자/사망자 별 평균으로 대체하기

```python
mean_age = train_data['Age'].mean()
train_data['Age'].fillna(mean_age).head() ## inplace=True로 설정하면 데이터 실제로 변화

# 결측치를 열의 평균값으로 넣어줌
```

```python
# 생존자 나이 평균
alive_mean = train_data[train_data['Survived'] == 1]['Age'].mean()

# 사망자 나이 평균
dead_mean = train_data[train_data['Survived'] == 0]['Age'].mean()

print(alive_mean, dead_mean)
```

```python
# alive_mean으로 생존자중 나이가 NaN인 것들 채우기
# df.loc[행인덱싱, 열인덱싱]
train_data.loc[train_data['Survived'] == 1, 'Age'] = train_data[train_data['Survived'] == 1]['Age'].fillna(alive_mean)

# dead_mean으로 사망자중 나이가 NaN인 것들 채우기
train_data.loc[train_data['Survived'] == 0, 'Age'] = train_data[train_data['Survived'] == 0]['Age'].fillna(dead_mean)
```
